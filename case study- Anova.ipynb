```python
import pandas as pd
import numpy as np
```

```python
group1 = [85,86,88,75,78,94,98,79,71,80]
group2 = [91,92,93,85,87,84,82,88,95,96]
group3 = [79,78,88,94,92,85,83,85,82,81]
```

```python
from scipy.stats import f_oneway
```

```python
f_stat,p_value = f_oneway(group1,group2,group3)
```

```python
p_value
```

```python
if p_value<0.05:
    print('Reject null Hypothesis')
else:
    print('Accept null Hypothesis')
```

## Two way Anova

```python
df2 = pd.read_csv('Anova2.csv')
```

```python
df2
```

```python
df_melt = pd.melt(df2, id_vars = ['Genotype'], value_vars = ['1_year','2_year','3_year'])
```

```python
df_melt
```

```python
df_melt.columns = ['Genotype','Years','Value']
```

```python
df_melt.head()
```

```python
model1 = ols('Value ~ C(Genotype) + C(Years) + C(Genotype):C(Years)', data = df_melt).fit()
```

```python
anova_table = sm.stats.anova_lm(model1, typ = 2)
```

```python
anova_table
```

```python
df3 = pd.read_csv('Anova2way.csv')
```

```python
df3
```

```python
df_melt1 = pd.melt(df3, id_vars = ['Class'], value_vars = ['4-8 years','8-13 years','13-17 years'])
```

```python
df_melt1.columns = ['Class','Age_Group','Scores']
```

```python
df_melt1.head()
```
