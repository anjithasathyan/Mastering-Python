```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('scottish_hills.csv')
```

```python
x = df.Height
y = df.Latitude
```

```python
plt.scatter(x,y)
plt.xlabel('Height')
plt.ylabel('Latitude')
plt.title('Scatter plot of Height vs Latitude')
plt.show()
```

```python
x = df.Height
y = df.Longitude
```

```python
plt.scatter(x,y)
plt.xlabel('Height')
plt.ylabel('Longitude')
plt.title('Scatter plot of Height vs Longitude')
plt.show()
```

```python
plt.hist(df['Height'],bins=20,edgecolor='k', color = 'r') ## bins- how many ranges or divisions
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Histogram of Hill Height')
plt.show()
```

```python
plt.bar(df['Hill Name'],df['Height'])
plt.xlabel('Hill Name')
plt.ylabel('Height')
plt.title('Bar plot of Hill Heights by Hill Name')
plt.xticks(rotation=45)  ## slant to the x-labels by rotation
plt.show()
```

```python
## sort the dataset by height in descending order and select top ten
top_ten = df.sort_values(by='Height', ascending=False).head(10)

## create a line plot
plt.figure(figsize=(10,6))
for index, row in top_ten.iterrows():
    plt.plot(row['Osgrid'], row['Height'], marker='o',label=row['Hill Name'])
    
plt.xlabel('Osgrid')
plt.ylabel('Height')
plt.title('Top Ten Hill Stations by Height Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
small_ten = df.sort_values(by='Height', ascending=True).head(10)

## create a line plot
plt.figure(figsize=(10,6))
for index, row in small_ten.iterrows():
    plt.plot(row['Osgrid'], row['Height'], marker='o',label=row['Hill Name'])
    
plt.xlabel('Osgrid')
plt.ylabel('Height')
plt.title('small Ten Hill Stations by Height Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Latitude'],bins=20,edgecolor='k', color = 'b') ## bins- how many ranges or divisions
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Histogram of Hill Latitude')
plt.show()
```
