# univariate analysis of numerical columns

```python
# 1. Distplot
# 2.KDEplot
# 3.Boxplot
# 4.Violinplot
```

```python
# 1. distplot: histogram + PDF- probability distribution
# syntax:  sns.distplot(df.column_name,bins=__, color = 'r')

sns.distplot(data.total_bill, bins=30, color='g')
```

```python
# KDEplot ---> density

sns.kdeplot(data.total_bill, color='r')
```

```python
# distplot without KDE function---> only histogram
#sns.distplot(df.column_name,kde=False)

sns.distplot(data.total_bill,kde=False)
```

```python
#Boxplot
#sns.boxplot(x='col_name',data=df_name, color='b')

sns.boxplot(x='total_bill',data=data)
```

```python
# violinplot
#sns.violinplot(df.column_name)

sns.violinplot(data.total_bill)
```

```python
# 4 plots for univariate analysis on numerical column: total_bill
#num col: tip
```
